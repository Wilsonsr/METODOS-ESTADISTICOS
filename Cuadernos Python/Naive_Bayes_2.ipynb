```python
X = dataset.iloc[:, [1, 2, 3]].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
```

```python
X
```

### Dividir el conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Escala de características

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
X_test
```

-  Entrenamiento del modelo Naive Bayes en el conjunto de entrenamiento

```python
from sklearn.naive_bayes import GaussianNB
modelo = GaussianNB()
modelo.fit(X_train, y_train)

```

### Predicción de los resultados del conjunto de pruebas

```python
y_pred = modelo.predict(X_test)
```

```python
y_pred
```

```python
y_pred.size
```

- La matriz de confusión

```python
from sklearn.metrics import confusion_matrix, accuracy_score
ac = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True, cmap="Blues"  )
```

```python
ac # acuraccy
```

```python
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=modelo.classes_)
disp.plot()

plt.show()
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score, classification_report
```

```python
recall_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```
