- Máquinas de Vector Soporte (Vector Support Machines, SVMs) es un algoritmo de clasificación y regresión desarrollado en la década de los 90, dentro del campo de la ciencia computacional.

- Se desarrolló como un método de clasificación binaria


- Las Máquinas de Vector Soporte se fundamentan en el Maximal Margin Classifier, que a su vez, se basa en el concepto de hiperplano. 

Conceptualmente, los SVM son más fáciles de explicar para problemas de clasificación.

- ¿comprará el cliente este producto? [sí, no]
- ¿tipo de tumor? [maligno, benigno]
- ¿subirá el índice bursátil? IBEX mañana [sí, no]
- ¿es este comportamiento una anomalía? [sí, no]
- ¿nos devolverá este cliente un crédito? [sí, no]
- ¿qué deporte estás haciendo? tal y como lo detectan los relojes inteligentes [caminar, correr, bicicleta, nadar]
- ¿obtendrá una historia un número alto de visitas en un agregador de noticias? [sí, no]

- ## Vectores de apoyo

Los vectores de soporte son los puntos de datos que están más cerca del hiperplano. Estos puntos definirán mejor la línea de separación al calcular los márgenes. Estos puntos son más relevantes para la construcción del clasificador.

- ## Hiperplano
Un hiperplano es un plano de decisión que separa un conjunto de objetos que pertenecen a diferentes clases.

- ## Margen

El margen es un espacio entre las dos líneas en los puntos de clase más cercanos. 
  -Se calcula como la distancia perpendicular desde la línea hasta los vectores de apoyo o los puntos más cercanos. 
  - Si el margen es mayor entre las clases, entonces se considera un buen margen, un margen más pequeño es un mal margen.

# SVM (Support Vector Machines)

Las maquinas de soporte vectorial, maquinas de vector soporte es un algoritmo inicialmente construido para clasficaciones más robustas que las realizadas con árboles. 
El algoritmo detallado está [acá](https://www.csie.ntu.edu.tw/~cjlin/papers/libsvm.pdf)

Lo primero que se construye es un hiperplano de separación, este es de la forma
$$\beta_1 X_1 + \ldots + \beta_p X_p=k$$
buscando que todos los puntos de una categoría estén en una sola parte del hiperplano.

Cómo esto es dificil de lograr, se trabaja con una función de coste C, de tal manera que a mayor valor de C, menos puntos quedan mal clasificados, pero se necesitan más hiperplanos. 





# Kernels

Para aumentar la dimensión de las variables y construir funciones más complejas se pueden usar
1. Lineal 
$$K(x,y)=x\cdot y$$
2. Polinomico

$$K(x,y)=(x\cdot y+a)^d$$
No se recomienda d>5 por sobreajuste

3. Gaussiana (RBF)

$$K(x,y)=e^{-\gamma ||x-y||^2}$$


Para más información, [click](http://crsouza.com/2010/03/17/kernel-functions-for-machine-learning-applications/#sigmoid)


## El truco del kernel en SVM

En ocasiones en las que no hay forma de encontrar una hiperplano que permita separar dos clases. En estos casos decimos que las clases no son linealmente separables.


- Para resolver este problema  usar el truco del kernel:

El truco del kernel consiste en inventar una dimensión nueva en la que podamos encontrar un hiperplano para separar las clases. En la siguiente figura vemos cómo al añadir una dimensión nueva, podemos separar fácilmente las dos clases con una superficie de decisión.

## Las ventajas de las máquinas de vectores de soporte son:

- Eficaz en espacios de altas dimensiones.

Todavía efectivo en casos donde el número de dimensiones es mayor que el número de muestras.

- Utiliza un subconjunto de puntos de entrenamiento en la función de decisión (llamados vectores de soporte), por lo que también es eficiente en memoria.

- Versátil: se pueden especificar diferentes funciones del Kernel para la función de decisión. Se proporcionan kernels comunes, pero también es posible especificar kernels personalizados.

## Las desventajas de las máquinas de vectores de soporte incluyen:

- Si el número de funciones es mucho mayor que el número de muestras, es crucial evitar el ajuste excesivo al elegir las funciones del núcleo y el término de regularización.

- Las SVM no proporcionan directamente estimaciones de probabilidad, estas se calculan mediante una costosa validación cruzada de cinco veces .
